```python
for i in range(4):
    for j in range(i + 1, 5):
        print('[i = {}, j = {}] '.format(i, j), end='')
    print()
```

```python
data = [8, 3, 4, 9, 1]
for i in range(4):
    for j in range(i + 1, 5):
        #   => (i)  (j)      .
        if data[i] > data[j]:
            data[i], data[j] = data[j], data[i]
        # ===== if
    # for j =>  
    print('{} : {}'.format(i + 1, data))
# for i =>  
# print(' : {}'.format(data))
```

```python
data = [8, 3, 4, 9, 1]
for i in range(4):
    for j in range(i + 1, 5):
        #   => (i)  (j)      .
        #  '>'    '<'   .
        if data[i] < data[j]:
            data[i], data[j] = data[j], data[i]
        # ===== if
    # for j =>  
    print('{} : {}'.format(i + 1, data))
# for i =>  
# print(' : {}'.format(data))
```

```python
data = [8, 3, 4, 9, 1]
for i in range(len(data) - 1): # (n - 1),   
    for j in range(i + 1, len(data)): #       
        if data[i] > data[j]:
            data[i], data[j] = data[j], data[i]
    print('{} : {}'.format(i + 1, data))
# print(' : {}'.format(data))
```
